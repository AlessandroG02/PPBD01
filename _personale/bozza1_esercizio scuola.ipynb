```python
import pandas as pd
import shutil

# imposta i percorsi dei file di input
file1_path = 'percorso/file1.xlsx'
file2_path = 'percorso/file2.xlsx'

# imposta il percorso del file di output
output_path = 'percorso/nuovo_file.xlsx'

# legge i dati dai file Excel
df1 = pd.read_excel(file1_path)
df2 = pd.read_excel(file2_path)

# unisce i dati in base alla colonna di confronto
merged_df = pd.merge(df1, df2, on='colonna_di_confronto')

# filtra i dati in base alla colonna di interesse
filtered_df = merged_df[merged_df['colonna_di_interesse'] == True]

# itera attraverso le righe del DataFrame filtrato
for index, row in filtered_df.iterrows():
    # legge il percorso del file da una colonna del DataFrame
    file_path = row['colonna_del_percorso_del_file']
    # ottiene il nome del file dal percorso
    file_name = file_path.split('/')[-1]
    # copia il file in un nuovo percorso
    shutil.copy(file_path, 'percorso_output/' + file_name)

# scrive i dati filtrati in un nuovo file Excel
filtered_df.to_excel(output_path, index=False)
```

```python
import tkinter as tk
from tkinter import filedialog

window = tk.Tk()

def seleziona_file1():
    file1 = filedialog.askopenfilename(filetypes=[("File Excel", "*.xlsx")])
    if file1:
        file1_label.config(text=f"File selezionato: {file1}")

def seleziona_file2():
    file2 = filedialog.askopenfilename(filetypes=[("File Excel", "*.xlsx")])
    if file2:
        file2_label.config(text=f"File selezionato: {file2}")

seleziona_file1_button = tk.Button(window, text="Seleziona file 1", command=seleziona_file1)
file1_label = tk.Label(window, text="Nessun file selezionato")

seleziona_file2_button = tk.Button(window, text="Seleziona file 2", command=seleziona_file2)
file2_label = tk.Label(window, text="Nessun file selezionato")

seleziona_file1_button.grid(row=2, column=0, padx=10, pady=10)
file1_label.grid(row=2, column=1, padx=10, pady=10, columnspan=2)

seleziona_file2_button.grid(row=3, column=0, padx=10, pady=10)
file2_label.grid(row=3, column=1, padx=10, pady=10, columnspan=2)

intervallo_celle_1_label = tk.Label(window, text="Intervallo celle 1:")
intervallo_celle_1 = tk.Entry(window, width=10)
intervallo_celle_2_label = tk.Label(window, text="Intervallo celle 2:")
intervallo_celle_2 = tk.Entry(window, width=10)

intervallo_righe_1_label = tk.Label(window, text="Intervallo righe 1:")
intervallo_righe_1 = tk.Entry(window, width=10)
intervallo_righe_2_label = tk.Label(window, text="Intervallo righe 2:")
intervallo_righe_2 = tk.Entry(window, width=10)

intervallo_colonne_1_label = tk.Label(window, text="Intervallo colonne 1:")
intervallo_colonne_1 = tk.Entry(window, width=10)
intervallo_colonne_2_label = tk.Label(window, text="Intervallo colonne 2:")
intervallo_colonne_2 = tk.Entry(window, width=10)

confronta_button = tk.Button(window, text="Confronta", command=confronta_righe_colonne)

output_celle_label = tk.Label(window, text="Valore in finestra Label output per intervallo_celle_1 vs intervallo_celle_2")
output_riga_label = tk.Label(window, text="Valore vero in finestra Label output per riga 1 input vs riga 2 input")

intervallo_celle_1_label.grid(row=4, column=0, padx=10, pady=10, sticky="W")
intervallo_celle_1.grid(row=4, column=1, padx=10, pady=10)
intervallo_celle_2_label.grid(row=4, column=2, padx=10, pady=10, sticky="W")
intervallo_celle_2.grid(row=4, column=3, padx=10, pady=10)

intervallo_righe_1_label.grid(row=1, column=0, padx=10, pady=10, sticky="W")
intervallo_righe_1.grid(row=1, column=1, padx=10, pady=10)
intervallo_righe_2_label.grid(row=1, column=2, padx=10, pady=10, sticky="W")
intervallo_righe_2.grid(row=1, column=3, padx=10, pady=10)

intervallo_colonne_1_label.grid(row=3, column=2, padx=10, pady=10, sticky="W")
intervallo_colonne_1.grid(row=3, column=3, padx=10, pady=10)
intervallo_colonne_2_label.grid(row=3, column=4, padx=10, pady=10, sticky="W")
intervallo_colonne_2.grid(row=3, column=5, padx=10, pady=10)

confronta_button.grid(row=5, column=0, padx=10, pady=10)

output_celle_label.grid(row=6, column=0, padx=10, pady=10, columnspan=2)
output_riga_label.grid(row=6, column=2, padx=10, pady=10, columnspan=2)
import openpyxl

def confronta_righe_colonne():
    file1 = file1_label.cget("text").split(": ")[1]
    file2 = file2_label.cget("text").split(": ")[1]

    if file1 and file2:
        wb1 = openpyxl.load_workbook(file1)
        wb2 = openpyxl.load_workbook(file2)

        sheet1 = wb1.active
        sheet2 = wb2.active

        intervallo_celle1 = intervallo_celle_1.get()
        intervallo_celle2 = intervallo_celle_2.get()
        intervallo_righe1 = intervallo_righe_1.get()
        intervallo_righe2 = intervallo_righe_2.get()
        intervallo_colonne1 = intervallo_colonne_1.get()
        intervallo_colonne2 = intervallo_colonne_2.get()

        valore_celle1 = sheet1[intervallo_celle1].value
        valore_celle2 = sheet2[intervallo_celle2].value

        righe1 = sheet1[intervallo_righe1]
        righe2 = sheet2[intervallo_righe2]

        valori_riga1 = [cella.value for cella in righe1]
        valori_riga2 = [cella.value for cella in righe2]

        colonne1 = sheet1[intervallo_colonne1]
        colonne2 = sheet2[intervallo_colonne2]

        valori_colonna1 = [cella.value for cella in colonne1]
        valori_colonna2 = [cella.value for cella in colonne2]

        valore_celle_output = ""
        valore_riga_output = ""

        if valore_celle1 == valore_celle2:
            valore_celle_output = f"I valori delle celle {intervallo_celle1} e {intervallo_celle2} sono uguali"
        else:
            valore_celle_output = f"I valori delle celle {intervallo_celle1} e {intervallo_celle2} sono diversi"

        if valori_riga1 == valori_riga2:
            valore_riga_output = f"I valori delle righe {intervallo_righe1} e {intervallo_righe2} sono uguali"
        else:
            valore_riga_output = f"I valori delle righe {intervallo_righe1} e {intervallo_righe2} sono diversi"

        output_celle_label.config(text=valore_celle_output)
        output_riga_label.config(text=valore_riga_output)

window.mainloop()

```
